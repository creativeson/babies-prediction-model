```python
X = df[['marriage_couples', 'female_marriage_age','male_marriage_age']]
```

```python
y = df[['total_birth_numbers']]
y
```

# Use linear algebra to get linear regresion model

```python
X["intercept"] = 1
```

```python
X = X [["intercept", "marriage_couples", "female_marriage_age",'male_marriage_age']]
```

```python
X
```

```python
X_T =  X.T
```

```python
X_T
```

## y = XB
## X_T * y = X_T * X * B
## (X_TX)^-1 * X_T * y = (X_TX)^-1 * X_T * X * B
## B = (X_T*X)^-1 * X_T * y


```python
B = np.linalg.inv(X_T @ X) @ X_T @ y 
```

```python
X
```

```python
B.index = X.columns
```

```python
B
```

```python
predictions = X @ B
predictions
```

## The R2 statistic measures the accuracy of my prediction against the actual numbers.It is close to one when the errors between my prediction and the actual numbers are relatively small.

```python
SSR = ((y - predictions)**2).sum() 
SST = ((y - y.mean())**2).sum()
```

```python
R2 = 1 - (SSR /SST)
R2
```
